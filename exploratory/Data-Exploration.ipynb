```python
### clean and vectorise - Question Title
from sklearn.feature_extraction.text import TfidfVectorizer
tf = TfidfVectorizer(min_df=3, stop_words='english')
title = tf.fit_transform(df['Title'])
```

```python
# expect a matrix / sparse matrix with 1 row per record (stackoverflow question) and 1 column per term/word feature
# word features have values equal to term-inverse frequency of terms in each record
# these will be modelled against the open/close rate of each record
# e.g. in regression, a coefficient will be fitted against each word feature
# in a decision tree, partitions of word features will be found to split records into open/close predictions
title
```

```python
# check no. rows of TF-IDF output
title.shape[0] == df.shape[0]
```
